```python
import pandas as pd
Merged_Walmart = pd.read_excel(r"C:\Users\adishree.b\Documents\Creatives_New.xlsx",Sheetname ='Walmart')
```

```python
Merged_Walmart.replace('NaN','Others',inplace=True)
```

```python
boxplot = Merged_Walmart.boxplot(column=['Impressions'], by = 'Targeting Type', figsize= (20,16))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(20,5))
plt.xlabel('Year of registration')
plt.ylabel('Number of Vehicles registered')
plt.title('Distribution of vehicles by year of Registration')
Merged_Walmart['Targeting Type'].value_counts().plot.bar()
```

```python
fig = plt.figure(figsize=(20,5))
plt.xlabel('Year of registration')
plt.ylabel('Number of Vehicles registered')
plt.title('Distribution of vehicles by year of Registration')
Merged_Walmart.groupby(['Targeting Type']).Impressions.mean().plot.bar()
```

```python
fig = plt.figure(figsize=(20,5))
plt.xlabel('Year of registration')
plt.ylabel('Number of Vehicles registered')
plt.title('Distribution of vehicles by year of Registration')
Merged_Walmart.groupby(['Creative Size']).Impressions.mean().plot.bar()
```

```python
fig = plt.figure(figsize=(20,5))
plt.xlabel('Year of registration')
plt.ylabel('Number of Vehicles registered')
plt.title('Distribution of vehicles by year of Registration')
Merged_Walmart.groupby(['Creative Size','Targeting Type']).Impressions.mean().plot.bar(subplots=True)
```
